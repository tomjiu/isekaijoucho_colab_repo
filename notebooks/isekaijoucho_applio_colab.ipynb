Fill the GitHub repository URL and Release asset URLs, then run all cells.

```python
#@title 1. Settings
REPO_URL = "https://github.com/tomjiu/isekaijoucho_colab_repo"  #@param {type:"string"}
DATA_ZIP_URL = "https://github.com/tomjiu/isekaijoucho_colab_repo/releases/download/v0-data/isekaijoucho_source_audio.zip"  #@param {type:"string"}
CHECKPOINT_ZIP_URL = "https://github.com/tomjiu/isekaijoucho_colab_repo/releases/download/v0-data/isekaijoucho_epoch1_checkpoint.zip"  #@param {type:"string"}
PREPROCESSED_ZIP_URL = "https://github.com/tomjiu/isekaijoucho_colab_repo/releases/download/v0-data/isekaijoucho_preprocessed_logs.zip"  #@param {type:"string"}
MODEL_NAME = "isekaijoucho_whispertrim"  #@param {type:"string"}
TOTAL_EPOCH = 20  #@param {type:"integer"}
BATCH_SIZE = 6  #@param {type:"integer"}
F0_METHOD = "rmvpe"  #@param ["rmvpe", "crepe-tiny", "crepe", "fcpe"]
GITHUB_REPO = "tomjiu/isekaijoucho_colab_repo"  #@param {type:"string"}
GITHUB_BACKUP_TAG = "colab-model-backups"  #@param {type:"string"}

assert DATA_ZIP_URL, "Paste the GitHub Release URL for isekaijoucho_source_audio.zip"
```

```python
#@title 1.5 GitHub backup token
# Create a fine-grained GitHub token with Contents: Read and write for this repository only.
# Leave empty to train without automatic GitHub Release backups.
import getpass
import os
GITHUB_TOKEN = getpass.getpass("GitHub token for automatic model backups: ")
os.environ["GITHUB_TOKEN"] = GITHUB_TOKEN
```

```python
#@title 4. Train
!python /content/isekaijoucho_colab_repo/scripts/colab_train.py \
  --repo-dir /content/isekaijoucho_colab_repo \
  --applio-dir /content/Applio \
  --data-url "$DATA_ZIP_URL" \
  --checkpoint-url "$CHECKPOINT_ZIP_URL" \
  --preprocessed-url "$PREPROCESSED_ZIP_URL" \
  --model-name "$MODEL_NAME" \
  --total-epoch "$TOTAL_EPOCH" \
  --batch-size "$BATCH_SIZE" \
  --f0-method "$F0_METHOD" \
  --github-repo "$GITHUB_REPO" \
  --github-tag "$GITHUB_BACKUP_TAG"
```
